```python
import pandas as pd
from sqlalchemy import create_engine
import os

# MySQL connection setup
username = 'root'
password = 'suinlmala'
host = 'localhost'
port = 3306
database = 'sunil'

engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}')

data_folder = '.'

for file in os.listdir(data_folder):
    if file.endswith('.csv'):
        try:
            file_path = os.path.join(data_folder, file)
            table_name = os.path.splitext(file)[0]
            print(f"üìÇ Processing {file}")

            # Read in chunks
            chunk_iter = pd.read_csv(file_path, chunksize=10000)
            for i, chunk in enumerate(chunk_iter):
                # Clean column names
                chunk.columns = [col.strip().replace(' ', '_').lower() for col in chunk.columns]
                # Append or replace based on chunk number
                if_exists = 'replace' if i == 0 else 'append'
                chunk.to_sql(name=table_name, con=engine, if_exists=if_exists, index=False)

            print(f"‚úÖ Uploaded: {file} as table `{table_name}`")

        except Exception as e:
            print(f"‚ùå Failed to upload {file}: {e}")
```

```python
import numpy as np

# Replace inf, -inf with NaN, then fill NaN with 0 or None
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)  # or use df.fillna(None, inplace=True) for NULL in SQL
```

```python
df=pd.read_sql("""select * from final_sale_table_updated
where gross_profit>0
and profit_margin>0
and totalsalequantity>0""",con=engine)
```

```python
pd.read_sql("""select brand,totalsaledollars,profit_margin from final_sale_table_updated
where totalsaledollars>0
and profit_margin>0
order by totalsaledollars asc""",con=engine)
```

```python
brand_performance=df.groupby("description").agg({
    "totalsaledollars":"sum",
    "profit_margin":"mean"
}
).reset_index()
```

```python
low_sale_threshold=brand_performance["totalsaledollars"].quantile(0.15)
high_profit_threshold=brand_performance["profit_margin"].quantile(0.85)
```

```python
print(low_sale_threshold)
print(high_profit_threshold)
```

```python
target_brands=brand_performance[(brand_performance["totalsaledollars"] < low_sale_threshold) &
                                  (brand_performance["profit_margin"] > high_profit_threshold)]
display(target_brands.sort_values(by="totalsaledollars"))
```
